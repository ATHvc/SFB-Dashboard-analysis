```python
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="SFB Financial Dashboard", layout="wide")

st.title("Small Finance Bank Financial Dashboard")

df = pd.read_csv("cleaned_sfb_data.csv")

# Clean text
df["Bank"] = df["Bank"].astype(str).str.strip()
df["Metric"] = df["Metric"].astype(str).str.strip()

# Sidebar filters
st.sidebar.header("Filters")

selected_banks = st.sidebar.multiselect(
    "Select Banks",
    options=sorted(df["Bank"].unique()),
    default=sorted(df["Bank"].unique())
)

selected_metric = st.sidebar.selectbox(
    "Select Metric",
    options=sorted(df["Metric"].unique())
)

filtered = df[
    (df["Bank"].isin(selected_banks)) &
    (df["Metric"] == selected_metric)
].copy()

tab1, tab2, tab3, tab4 = st.tabs(
    ["Overview", "Bank Comparison", "Rankings", "Raw Data"]
)

with tab1:
    st.subheader("Dashboard Overview")

    col1, col2, col3, col4 = st.columns(4)

    col1.metric("Selected Banks", len(selected_banks))
    col2.metric("Selected Metric", selected_metric)
    col3.metric("Average Value", round(filtered["Value"].mean(), 4))
    col4.metric("Highest Value", round(filtered["Value"].max(), 4))

    st.write("This dashboard compares Small Finance Banks across key financial ratios and performance metrics.")

    fig = px.bar(
        filtered,
        x="Bank",
        y="Value",
        color="Bank",
        title=f"{selected_metric} Comparison"
    )
    st.plotly_chart(fig, use_container_width=True)

with tab2:
    st.subheader("Bank Comparison")

    chart_type = st.radio(
        "Select Chart Type",
        ["Bar Chart", "Line Chart"],
        horizontal=True
    )

    if chart_type == "Bar Chart":
        fig = px.bar(
            filtered,
            x="Bank",
            y="Value",
            color="Bank",
            title=f"{selected_metric} - Bank Wise Comparison"
        )
    else:
        fig = px.line(
            filtered,
            x="Bank",
            y="Value",
            markers=True,
            title=f"{selected_metric} Trend Across Banks"
        )

    st.plotly_chart(fig, use_container_width=True)

    st.subheader("Metric Heatmap")

    heatmap_df = df[df["Bank"].isin(selected_banks)].pivot_table(
        index="Bank",
        columns="Metric",
        values="Value",
        aggfunc="mean"
    )

    fig_heatmap = px.imshow(
        heatmap_df,
        text_auto=True,
        aspect="auto",
        title="Bank vs Metric Heatmap"
    )

    st.plotly_chart(fig_heatmap, use_container_width=True)

with tab3:
    st.subheader("Rankings")

    ranking_df = filtered.sort_values("Value", ascending=False)

    col1, col2 = st.columns(2)

    with col1:
        st.write("Top Banks")
        st.dataframe(ranking_df.head(5), use_container_width=True)

    with col2:
        st.write("Bottom Banks")
        st.dataframe(ranking_df.tail(5), use_container_width=True)

    fig_rank = px.bar(
        ranking_df,
        x="Value",
        y="Bank",
        orientation="h",
        color="Bank",
        title=f"Ranking by {selected_metric}"
    )
    st.plotly_chart(fig_rank, use_container_width=True)

with tab4:
    st.subheader("Raw Data")

    st.dataframe(df, use_container_width=True)

    csv = df.to_csv(index=False).encode("utf-8")

    st.download_button(
        label="Download Data as CSV",
        data=csv,
        file_name="sfb_financial_dashboard_data.csv",
        mime="text/csv"
    )
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="SFB Financial Dashboard", layout="wide")
st.title("Small Finance Bank Financial Dashboard")

df = pd.read_csv("cleaned_sfb_data.csv")
df["Bank"] = df["Bank"].astype(str).str.strip()
df["Metric"] = df["Metric"].astype(str).str.strip()

metric_notes = {
    "Revenue growth (YoY)": "Shows year-on-year growth in revenue.",
    "Gross margin": "Shows profitability after direct costs.",
    "EBITDA margin": "Shows operating profitability before interest, tax, depreciation, and amortization.",
    "PBT margin": "Shows profit before tax as a percentage of revenue.",
    "Debt equity ratio": "Shows financial leverage and debt dependency.",
    "ROA": "Shows how efficiently assets generate profit.",
    "ROE": "Shows return generated on shareholders’ equity.",
}

selected_banks = st.multiselect(
    "Select Banks",
    options=sorted(df["Bank"].unique()),
    default=sorted(df["Bank"].unique())
)

selected_metric = st.selectbox(
    "Select Metric",
    options=sorted(df["Metric"].unique())
)

filtered = df[
    (df["Bank"].isin(selected_banks)) &
    (df["Metric"] == selected_metric)
].copy()

filtered = filtered.sort_values("Value", ascending=False)

tab1, tab2, tab3, tab4 = st.tabs(["Overview", "Bank Comparison", "Rankings", "Raw Data"])

with tab1:
    st.subheader("Overview")

    st.info(metric_notes.get(selected_metric, "This metric is used to compare financial performance across banks."))

    top_bank = filtered.iloc[0]["Bank"] if not filtered.empty else "-"
    bottom_bank = filtered.iloc[-1]["Bank"] if not filtered.empty else "-"

    col1, col2, col3, col4 = st.columns(4)
    col1.metric("Selected Banks", len(selected_banks))
    col2.metric("Average Value", round(filtered["Value"].mean(), 4))
    col3.metric("Top Bank", top_bank)
    col4.metric("Bottom Bank", bottom_bank)

    fig = px.bar(
        filtered,
        x="Bank",
        y="Value",
        color="Bank",
        title=f"{selected_metric} Comparison"
    )
    st.plotly_chart(fig, use_container_width=True)

with tab2:
    st.subheader("Bank Comparison")

    fig = px.bar(
        filtered,
        x="Bank",
        y="Value",
        color="Bank",
        title=f"Bank-wise Comparison for {selected_metric}"
    )
    st.plotly_chart(fig, use_container_width=True)

    st.subheader("Metric Heatmap")

    heatmap_df = df[df["Bank"].isin(selected_banks)].pivot_table(
        index="Bank",
        columns="Metric",
        values="Value",
        aggfunc="mean"
    )

    fig_heatmap = px.imshow(
        heatmap_df,
        text_auto=True,
        aspect="auto",
        title="Bank vs Metric Heatmap"
    )
    st.plotly_chart(fig_heatmap, use_container_width=True)

with tab3:
    st.subheader("Rankings")

    col1, col2 = st.columns(2)

    with col1:
        st.write("Top Banks")
        st.dataframe(filtered.head(5), use_container_width=True)

    with col2:
        st.write("Bottom Banks")
        st.dataframe(filtered.tail(5).sort_values("Value"), use_container_width=True)

    fig_rank = px.bar(
        filtered,
        x="Value",
        y="Bank",
        orientation="h",
        color="Bank",
        title=f"Ranking by {selected_metric}"
    )
    st.plotly_chart(fig_rank, use_container_width=True)

with tab4:
    st.subheader("Raw Data")

    st.dataframe(df, use_container_width=True)

    csv = df.to_csv(index=False).encode("utf-8")

    st.download_button(
        label="Download CSV",
        data=csv,
        file_name="sfb_financial_dashboard_data.csv",
        mime="text/csv"
    )
```
